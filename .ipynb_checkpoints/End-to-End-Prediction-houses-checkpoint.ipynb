## Handling Missing Values and EDA

```python
target = train["SalePrice"].copy()
predictors = train.drop("SalePrice", axis=1)
```

```python
# Create Cluster similarity class

class ClusterSimilarity(BaseEstimator, TransformerMixin):
    def __init__(self, n_clusters=10, gamma=1., random_state=42):
        self.n_clusters = n_clusters
        self.gamma = gamma
        self.random_state= random_state
    
    def fit(self, X, y=None, sample_weight=None):
        """
            Fit Data into class. 
            X is dataset, y is target
            and return to self
        """
        self.kmeans_ = KMeans(self.n_clusters, random_state=self.random_state)
        self.kmeans_.fit(X, sample_weight=sample_weight)
        
        return self
    def transform(self, X):
        return rbf_kernel(X, self.kmeans_.cluster_centers_, gamma=self.gamma)
    
    def get_feature_names_out(self, names):
        return [f"Cluster {i}" for i in range(self.n_clusters)]
```

```python
cluster_simil = ClusterSimilarity(n_clusters=10, gamma=1., random_state=42)
similarity = cluster_simil.fit_transform(predictors[["OverallQual", "OverallCond","YearBuilt",
                                               "1stFlrSF", "2ndFlrSF", "GarageArea"]],
                                         sample_weight=target,)
```

```python
print(similarity)
```

```python
# Create lists for numerical, boolean and cateborical attributes

num_att = []
cat_att = []

for label, content in predictors.items():
    if pd.api.types.is_numeric_dtype(content):
        num_att.append(label)
    else:
        cat_att.append(label)
```

```python
print(f"Counts of numerical attributes - {len(num_att)}")
print(f"Counts of categorical&object attributes - {len(cat_att)}")
```

```python
# Prepare data - Handling missing values(Pipeline)
num_pipeline = Pipeline([
    ("num_impute", SimpleImputer(strategy="median")),
    ("num_scaler", StandardScaler())
])

cat_pipeline = Pipeline([
    ("cat_impute", SimpleImputer(strategy="most_frequent")),
    ("cat_scaler", OneHotEncoder(handle_unknown="ignore"))
])

cluster_simil = ClusterSimilarity(n_clusters=10, gamma=1., random_state=42)

default_num_pipeline = Pipeline([
    ("default_impute", SimpleImputer(strategy="median")),
    ("default_scaler", StandardScaler())
])
```

```python
# Transformation
    
preprocessing = ColumnTransformer([
    ("num", num_pipeline, num_att),
    ("cat", cat_pipeline, cat_att),
    ("similarity", cluster_simil, ["OverallQual", "OverallCond","YearBuilt",
                                "1stFlrSF", "2ndFlrSF", "GarageArea"])
], remainder=default_num_pipeline)
```

```python
# get Prepared data

prepared_train_data = preprocessing.fit_transform(predictors)
```

```python
prepared_train_data.shape
```

```python
pd.DataFrame(prepared_train_data, columns=preprocessing.get_feature_names_out())
```

## Select and train a model

```python
# Fit data with LinearRegressor
lin_reg = Pipeline([
    ("preprocessing", preprocessing),
    ("model", LinearRegression())
])
lin_reg.fit(predictors, target)
```

```python
lin_pred = lin_reg.predict(predictors)
pd.Series(lin_pred).describe()
```

```python
lin_rmse = mean_squared_error(target, lin_pred, squared=False)
lin_rmse
```

```python
# DecisionTreeRegressor

tree_reg = Pipeline([
    ("preprocessing", preprocessing),
    ("model", DecisionTreeRegressor(random_state=42))
])

tree_reg.fit(predictors, target)
tree_pred = tree_reg.predict(predictors)
```

```python
mean_squared_error(target, tree_pred)
```

```python
tree_rmses = -cross_val_score(tree_reg, predictors, target, scoring="neg_root_mean_squared_error", cv=10)
```

```python
pd.Series(tree_rmses).describe()
```

```python
# RandomForestRegressor

rnf_reg = Pipeline([
    ("preprocessing", preprocessing),
    ("model", RandomForestRegressor(random_state=42))
])

rnf_reg.fit(predictors, target)
```

```python
rnf_pred = rnf_reg.predict(predictors)
rnf_rmses = -cross_val_score(rnf_reg, predictors, target, scoring="neg_root_mean_squared_error", cv=10)
```

```python
pd.Series(rnf_rmses).describe()
```

## Fine-Tune Model

```python
full_pipline = Pipeline([
    ("preprocessing", preprocessing),
    ("random_forest", RandomForestRegressor(random_state=42))
])
```

```python
# GridSearchCV

param_grid = [
    {"preprocessing__similarity__n_clusters": [5,8,10],
    "random_forest__max_features": [4,6,8]},
    
    {"preprocessing__similarity__n_clusters": [10,15],
    "random_forest__max_features": [8,10,15]},
]

grid_cv = GridSearchCV(full_pipline, param_grid, cv=3, scoring="neg_root_mean_squared_error")
grid_cv.fit(predictors, target)
```

```python
grid_cv.best_params_
```

```python
cv_result = pd.DataFrame(grid_cv.cv_results_)
cv_result.sort_values(by="mean_test_score", ascending=False, inplace=True)
cv_result = cv_result[["param_preprocessing__similarity__n_clusters", "param_random_forest__max_features",
                "split0_test_score", "split1_test_score", "split2_test_score", "mean_test_score"]]
score_cols = ["split0", "split1","split2", "mean_test_rmse"]
cv_result.columns = ["n_clusters", "max_features"] + score_cols
cv_result[score_cols] = -cv_result[score_cols].round().astype(np.int64)
cv_result.head()
```

```python
# RandomizedSerachCV
from scipy.stats import randint
param_dist = {"preprocessing__similarity__n_clusters": randint(low=3, high=50),
                "random_forest__max_features": randint(low=2, high=250)}
rnd_cv = RandomizedSearchCV(
    full_pipline,
    param_distributions=param_dist,
    n_iter=10,
    cv=5,
    random_state=42,
    scoring="neg_root_mean_squared_error"
)
rnd_cv.fit(predictors, target)
```

```python
# Get final model
final_model = rnd_cv.best_estimator_
final_model
```

```python
feature_importances = final_model["random_forest"].feature_importances_
feature_importances.round(3)
```
